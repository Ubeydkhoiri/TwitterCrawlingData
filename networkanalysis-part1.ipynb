### Import Library

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import numpy as np
import re
```

### Twitter Full Data Crawling without API
We will take tweet data using keyword: **Ferdy Sambo** and **lang:id** (tweet in bahasa Indonesia only)

```python
query = '"Ferdy Sambo" lang:id'
tweets = []
limit = 10000

%time

try:
    print('Start crawling')
    for tweet in sntwitter.TwitterSearchScraper(query=query).get_items():
        if len(tweets) == limit:
            break
        else:
            if tweet.inReplyToUser is not None:
                tweet_type = 'reply'
                tweet_reply = re.findall(r'[/]\w+', str(tweet.inReplyToUser))[-1].replace('/','')
            elif tweet.quotedTweet is not None:
                tweet_type = 'retweet'
                tweet_reply = None
            else:
                tweet_type = 'original'
                tweet_reply = None
            tweets.append([tweet.conversationId,
                           tweet.coordinates,
                           tweet.date,
                           tweet.hashtags,
                           tweet.id,
                           tweet.inReplyToTweetId,
                           tweet.inReplyToUser,
                           tweet_reply,
                           tweet.lang,
                           tweet.likeCount,
                           tweet.media,
                           tweet.mentionedUsers,
                           tweet.outlinks,
                           tweet.place,
                           tweet.quoteCount,
                           tweet.quotedTweet,
                           tweet.renderedContent,
                           tweet_type,
                           tweet.replyCount,
                           tweet.retweetCount,
                           tweet.retweetedTweet,
                           tweet.sourceLabel,
                           tweet.url,
                           tweet.user,
                           tweet.user.username])

    df = pd.DataFrame(tweets, columns=(['conversationId',
                                        'coordinates',
                                        'date',
                                        'hashtags',
                                        'id',
                                        'inReplyToTweetId',
                                        'inReplyToUser_url',
                                        'inReplyToUser',
                                        'lang',
                                        'likeCount',
                                        'media',
                                        'mentionedUsers',
                                        'outlinks',
                                        'place',
                                        'quoteCount',
                                        'quotedTweet',
                                        'description',
                                        'tweet_type',
                                        'replyCount',
                                        'retweetCount',
                                        'retweetedTweet',
                                        'source',
                                        'tweet_url',
                                        'user_url',
                                        'username']))
except Exception as e:
    print(e)
    
print('Finished')
print('-----')
```

### Filter Data Tweet (Tweet Type = Reply)

```python
df_reply = df[['inReplyToUser','username']][df['tweet_type'] == 'reply']
df_reply
```

```python
# Sort the cases with a->b and b->a
relationship_df = pd.DataFrame(np.sort(df_reply.values, axis = 1), columns = ['from','to'])
relationship_df
```

```python
relationship_df["value"] = 1
relationship_df = relationship_df.groupby(["from","to"], sort=False, as_index=False).sum()
```
